## Train model

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)

print(x.shape, y.shape)
```

```python
scaler = StandardScaler().fit(x)
scaled_x = scaler.transform(x)

x_dev, x_test, y_dev, y_test = train_test_split(scaled_x, y, test_size=0.3)
x_train, x_val, y_train, y_val = train_test_split(x_dev, y_dev, test_size=0.3)
```

```python
lr_model = LogisticRegression(solver="saga", tol=0.1).fit(x_train, y_train)
```

```python
print(f"Training accuracy: {lr_model.score(x_train, y_train):.5f}")
print(f"Validation accuracy: {lr_model.score(x_val, y_val):.5f}")
print(f"Test accuracy: {lr_model.score(x_test, y_test):.5f}")
```

```python
filename = "lr_model.pkl"
pickle.dump(lr_model, open(filename, "wb"))
```
